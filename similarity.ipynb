```python
dct = {}

for idx in dftot.index:
    
    dfinit= pd.DataFrame(dftot.loc[idx]).transpose() #select row of dataframe as dataframe
    df = (pd.Series([tuple(x) for x in dftot[coords][idx][0]])
            .reset_index().rename(columns={0:coords,"index":"point"})) #join coordinates as tuples 
    
    df =  df.join(dfinit.drop(coords,1), how='left').reset_index(level=0) #integrate into larger dataframe

    df[['lat','long']] = df[coords].apply(pd.Series) #create lon and lat columns
    df = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.long, df.lat)) #create shapely points
    generated_points= gpd.points_from_xy(df.long, df.lat)
    generated_line=LineString(generated_points)
    distance_delta = 0.0003
    distances = np.arange(0, generated_line.length, distance_delta)
    points = [generated_line.interpolate(distance) for distance in distances] + [generated_line.boundary[1]]
    dct['%s' % dftot['properties.onestop_id'][idx]] = points
```

```python
dist=[]
dct_points={}
dct_linestrings={}
df_buses_points=pd.DataFrame()

for sel_key in dct:

    blackdict={}
    blackdict={key: val for key, val in dct.items() if key != str(sel_key)}

    for bkkey in blackdict:

        pointlist=[]
        point_prev=dct[sel_key][0]

        for point in dct[sel_key]:
            
            for bkpoint in blackdict[bkkey]:

                if point.distance(bkpoint)*111000<20: # if distance < 20m
                        
                    pointlist.append(point)


        

        if pointlist:

            point_prev=pointlist[0]
            mylist1=[]
            list1=[]
            mylist2=[]
            list2=[]
            mylist3=[]
            list3=[]

            for point in pointlist:                
                
                if point.distance(point_prev)*111000>700:

                    if point.distance(point_prev2)*111000>700:
                        
                        list3.append(point)
                        mylist3=={'lines':'{key1} | {key2}'.format(key1=sel_key, key2=bkkey), 'geom':list3}

                    else:    

                        list2.append(point)
                        mylist2={'lines':'{key1} | {key2}'.format(key1=sel_key, key2=bkkey), 'geom':list2}
                        point_prev2=point
                
                else:

                    list1.append(point)
                    mylist1={'lines':'{key1} | {key2}'.format(key1=sel_key, key2=bkkey),'geom':list1}
                    point_prev=point
                    point_prev2=point
            

            df_buses_points=df_buses_points.append(mylist1, ignore_index=True)
            df_buses_points=df_buses_points.append(mylist2, ignore_index=True)
            df_buses_points=df_buses_points.append(mylist3, ignore_index=True)

     
    

```
