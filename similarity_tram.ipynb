```python
dct = {}

for idx in dftot.index:
    
    dfinit= pd.DataFrame(dftot.loc[idx]).transpose() #select row of dataframe as dataframe
    df = (pd.Series([tuple(x) for x in dftot[coords][idx][0]])
            .reset_index().rename(columns={0:coords,"index":"point"})) #join coordinates as tuples 
    
    df =  df.join(dfinit.drop(coords,1), how='left').reset_index(level=0) #integrate into larger dataframe

    df[['lat','long']] = df[coords].apply(pd.Series) #create lon and lat columns
    df = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.long, df.lat)) #create shapely points
    generated_points= gpd.points_from_xy(df.long, df.lat)
    generated_line=LineString(generated_points)
    distance_delta = 0.0003
    distances = np.arange(0, generated_line.length, distance_delta)
    points = [generated_line.interpolate(distance) for distance in distances] + [generated_line.boundary[1]]
    dct['%s' % dftot['properties.onestop_id'][idx]] = points
```

```python
# IDENTIFY AND DELETE COMMON TRAJECTORIES

lines_shapes = [[loads(l)] for l in df_buses_linestrings['geom']]
df_buses_linestrings['coords']=1 # preassign dummy value
def f(idx):
    xy = lines_shapes[idx][0].coords.xy
    longs = xy[0].tolist()
    lats = xy[1].tolist()
    return [list(z) for z in zip(lats, longs)]

for idx, row in df_buses_linestrings.iterrows():

    df_buses_linestrings.at[idx, 'coords'] = f(idx)

```

```python
index_list_del=[]
capstone_idx=[]

for idx1, row1 in df_buses_linestrings.iterrows():

    capstone_idx.append(idx1)  # keep track of indexes
    
    for idx2, row2 in df_buses_linestrings.iterrows():

        traj_A = np.array(row1["coords"])
        traj_B = np.array(row2["coords"])

        dist = tdist.sspd(traj_A, traj_B)

        # filtering out similar trajectories
        
        if idx2 not in capstone_idx:

            if dist < 9e-04: 
                
                #clean_df_buses_lnstr.drop(idx, axis=0, inplace=True)
                index_list_del.append(idx1)
                print(dist)
```
